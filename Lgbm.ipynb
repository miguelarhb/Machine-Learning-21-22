## Model - Lightgbm Optuna Hyperparameters Optimization

```python
print('Starting model...')

def objective(trial):  
    
    params = {
            'objective': 'regression',
            'boosting_type': 'goss',
            'metric': 'rmse',  
            'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.05),
            'other_rate': trial.suggest_loguniform('other_rate', 0.001, 0.4),
            'top_rate': trial.suggest_loguniform('top_rate', 0.01, 0.6),
            'max_depth': 20,
            'num_leaves' : trial.suggest_int('num_leaves', 300, 400),
            'n_estimators': trial.suggest_int('n_estimators', 4000, 9000),
            'reg_alpha': trial.suggest_loguniform('reg_alpha', 0.001, 10.0),
            'reg_lambda': trial.suggest_loguniform('reg_lambda', 0.001, 10.0),
            'colsample_bytree': 0.01,
            'subsample': trial.suggest_categorical('subsample', [0.4, 0.5, 0.6]),
            'min_child_samples': trial.suggest_int('min_child_samples', 100, 200)
    }
       
    model = LGBMRegressor(**params) 
    
    # avg_rmse=[]
    
    tss = TimeSeriesSplit(n_splits=3)
    for train_index, test_index in tss.split(X):
        train_x, test_x = X.iloc[train_index, :], X.iloc[test_index,:]
        train_y, test_y = y.iloc[train_index], y.iloc[test_index]
        
    model.fit(train_x,train_y,eval_set=[(test_x,test_y)],early_stopping_rounds=100,verbose=False)

    preds = model.predict(test_x)

    rmse = mean_squared_error(test_y, preds,squared=False)

    #    avg_rmse.append(rmse)
    
    return rmse

study = optuna.create_study(direction='minimize')

study.optimize(objective, n_trials=20)

print('Number of finished trials:', len(study.trials))
print('Best trial:', study.best_trial.params)
```
