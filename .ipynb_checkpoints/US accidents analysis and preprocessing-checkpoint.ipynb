```python
usAccidents = pd.read_csv("US_Accidents_Dec20_updated.csv") #Heavy
#usAccidents
```

## Data cleaning

There are no duplicates

```python
#Remove useless columns
#Country useless
usAccidents = usAccidents.drop(['Country'], axis=1)
#Bump have only around 0% true
usAccidents = usAccidents.drop(['Bump'], axis=1)
#Giveway have only around 0.2% true
usAccidents = usAccidents.drop(['Give_Way'], axis=1)
#No_Exit have only around 0.1% true
usAccidents = usAccidents.drop(['No_Exit'], axis=1)
#Roundabout have only around 0% true
usAccidents = usAccidents.drop(['Roundabout'], axis=1)
#Traffic_Calming have only around 0% true
usAccidents = usAccidents.drop(['Traffic_Calming'], axis=1)
#Turning_Loop have only around 0% true
usAccidents = usAccidents.drop(['Turning_Loop'], axis=1)
```

```python
# Cast Start_Time to datetime 

usAccidents["Start_Time"] = pd.to_datetime(usAccidents["Start_Time"])

# Extract year, month, weekday and day
usAccidents["Year"] = usAccidents["Start_Time"].dt.year
usAccidents["Month"] = usAccidents["Start_Time"].dt.month
usAccidents["Weekday"] = usAccidents["Start_Time"].dt.weekday
usAccidents['Dayofweek'] = usAccidents['Start_Time'].dt.dayofweek
usAccidents["Day"] = usAccidents["Start_Time"].dt.day
usAccidents['Week'] = usAccidents['Start_Time'].dt.isocalendar().week


# Extract hour and minute
usAccidents["Hour"] = usAccidents["Start_Time"].dt.hour
usAccidents["Minute"] = usAccidents["Start_Time"].dt.minute
```

```python
#ID is useless I guess
usAccidents = usAccidents.drop(['ID'], axis=1)
#ID is useless I guess
usAccidents = usAccidents.drop(['Description'], axis=1)
```

```python
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Thunder|T-Storm", na=False), "Weather_Condition"] = "Thunderstorm"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Snow|Sleet|Wintry", na=False), "Weather_Condition"] = "Snow"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Rain|Drizzle|Shower", na=False), "Weather_Condition"] = "Rain"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Wind|Squalls", na=False), "Weather_Condition"] = "Windy"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Hail|Pellets", na=False), "Weather_Condition"] = "Hail"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Fair", na=False), "Weather_Condition"] = "Clear"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Cloud|Overcast", na=False), "Weather_Condition"] = "Cloudy"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Mist|Haze|Fog", na=False), "Weather_Condition"] = "Fog"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Sand|Dust", na=False), "Weather_Condition"] = "Sand"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("Smoke|Volcanic Ash", na=False), "Weather_Condition"] = "Smoke"
usAccidents.loc[usAccidents["Weather_Condition"].str.contains("N/A Precipitation", na=False), "Weather_Condition"] = np.nan
```

```python
usAccidents.loc[usAccidents["Wind_Direction"] == "CALM", "Wind_Direction"] = "Calm"
usAccidents.loc[usAccidents["Wind_Direction"] == "VAR", "Wind_Direction"] = "Variable"
usAccidents.loc[usAccidents["Wind_Direction"] == "East", "Wind_Direction"] = "E"
usAccidents.loc[usAccidents["Wind_Direction"] == "North", "Wind_Direction"] = "N"
usAccidents.loc[usAccidents["Wind_Direction"] == "South", "Wind_Direction"] = "S"
usAccidents.loc[usAccidents["Wind_Direction"] == "West", "Wind_Direction"] = "W"

usAccidents["Wind_Direction"] = usAccidents["Wind_Direction"].map(lambda x : x if len(x) != 3 else x[1:], na_action="ignore")
```

```python
#More than 2/3 of the values in Number are Nan, the whole column can be discarded 
#(this data is not very useful for our analysis)
usAccidents = usAccidents.drop(['Number'], axis=1)
#Approximately 1/3 of the values in Wind_chill are Nan, the whole column can be discarded
#It is also higly related to Temperature
usAccidents = usAccidents.drop(['Wind_Chill(F)'], axis=1)

# EndLat and EndLng are highly correlated to Start, are not useful in our case
usAccidents = usAccidents.drop(['End_Lat','End_Lng'], axis=1)
usAccidents = usAccidents.drop(['End_Time'], axis=1)
usAccidents = usAccidents.drop(['Timezone','Airport_Code'], axis=1)
usAccidents = usAccidents.drop(['Weather_Timestamp'], axis=1)#Timestamp when is done observation
```

```python
#TODO maybe remove
#usAccidents['Precipitation(in)']= usAccidents['Precipitation(in)'].fillna(usAccidents['Precipitation(in)'].mean()) 

usAccidents.dropna(inplace=True)

usAccidents.isnull().sum()
```
